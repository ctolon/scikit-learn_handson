#### Verilerin yüklenmesi

```python
import pandas as pd

cv_results = pd.read_csv("dataset/randomized_search_results.csv", index_col=0)
cv_results
```

#### Prefix temizleyici ve cv sonuçlarını buna göre tekrar yapılandırma

```python
def shorten_param(param_name):
    if "__" in param_name:
        return param_name.rsplit("__", 1)[1]
    return param_name

cv_results = cv_results.rename(shorten_param, axis=1)
cv_results
```

#### Hyper parametrelerin 2 eksenli grafikte görselleştirme

Bu çizimde, en iyi performans gösteren değerlerin 0,01 ile 1,0 arasındaki bir öğrenme oranı bandında bulunduğunu görüyoruz, ancak diğer hiperparametrelerin öğrenme oranı için bu tür değerlerle nasıl etkileşime girdiği konusunda hiçbir kontrolümüz yok. Bunun yerine, paralel bir koordinat grafiği kullanarak tüm hiperparametreleri aynı anda görselleştirebiliriz.

```python
import seaborn as sns
import numpy as np

df = pd.DataFrame(
    {
        "max_leaf_nodes": cv_results["max_leaf_nodes"],
        "learning_rate": cv_results["learning_rate"],
        "score_bin": pd.cut(
            cv_results["mean_test_score"], bins=np.linspace(0.5, 1.0, 6)
        ),
    }
)
sns.set_palette("YlGnBu_r")
ax = sns.scatterplot(
    data=df,
    x="max_leaf_nodes",
    y="learning_rate",
    hue="score_bin",
    s=50,
    color="k",
    edgecolor=None,
)
ax.set_xscale("log")
ax.set_yscale("log")

_ = ax.legend(title="mean_test_score", loc="center left", bbox_to_anchor=(1, 0.5))
```

#### Paralel bir koordinat grafiği ile tüm hyper parametreleri görselleştirme

**Not:** Etkin aralıkları yaymak ve çizimin okunabilirliğini artırmak için bir log10 veya log2 alarak çoğu eksen değerini dönüştürdük.

Paralel koordinatlar grafiği, performans metriği renk kodluyken farklı sütunlardaki hiperparametrelerin değerlerini gösterecektir. Böylece, çalışan veya çalışmayan bir dizi hiperparametre olup olmadığını hızlı bir şekilde kontrol edebiliyoruz.

Paralel koordinat grafiğinin herhangi bir eksenine tıklayıp basılı tutarak bir sonuç aralığı seçmek mümkündür. Ardından aralık seçimini kaydırabilir (taşıyabilir) ve kesişmeleri görmek için iki seçim arasında geçiş yapabilirsiniz. Aynı eksene bir kez daha tıklayarak seçimi geri alabilirsiniz.

Özellikle bu hiperparametre araması için, sarı çizgilerin (en iyi performans gösteren modeller) hepsinin öğrenme oranı için ara değerlere, yani -2 ile 0 arasındaki kene değerlerine, yani 0,01 ila 1,0 öğrenme oranı değerlerine karşılık geldiğini doğrulamak ilginçtir. bir kez o eksen için log10 dönüşümünü tersine çevirdik.

Ancak artık max_bins ekseninde 1 ile 3 arasında tick değerleri olan satırları seçerek en yüksek performanslı modelleri seçmenin mümkün olmadığını da gözlemleyebiliyoruz.

Diğer hiperparametreler çok hassas değildir. Learning_rate eksen tick değerlerini -1.5 ile -0.5 arasında ve max_bins tick değerlerini 5 ile 8 arasında seçersek, diğer hiperparametrelerin değerleri ne olursa olsun her zaman en iyi performans gösteren modelleri seçtiğimizi kontrol edebiliriz.

Bu not defterinde, birden çok etkileşimli hiperparametre ile büyük bir rasgele aramanın sonuçlarını etkileşimli olarak nasıl keşfedeceğimizi gördük. Özellikle, bazı hiperparametrelerin çapraz doğrulama skoru üzerinde çok az etkiye sahip olduğunu, diğerlerinin ise iyi tahmin doğruluğuna sahip modeller elde etmek için belirli bir aralıkta ayarlanması gerektiğini gözlemledik.

```python
import numpy as np
import plotly.express as px

fig = px.parallel_coordinates(
    cv_results.rename(shorten_param, axis=1).apply(
        {
            "learning_rate": np.log10,
            "max_leaf_nodes": np.log2,
            "max_bins": np.log2,
            "min_samples_leaf": np.log10,
            "l2_regularization": np.log10,
            "mean_test_score": lambda x: x,
        }
    ),
    color="mean_test_score",
    color_continuous_scale=px.colors.sequential.Viridis,
)
fig.show()
```
